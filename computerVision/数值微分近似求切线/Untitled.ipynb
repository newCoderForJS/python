```python
import numpy as np
import matplotlib.pylab as plt
```

```python
def function_1(x):
    return 0.01*x**2 + 0.1*x
def numerical_diff(f, x):
    h = 1e-4 #0.0001
    return (f(x+h) - f(x-h)) / (2*h)
x = np.arange(0.0, 20.0, 0.1)#以0.1为单位, 从0到20的数组x
y = function_1(x)

plt.xlabel("x")
plt.ylabel("f(x)")

plt.plot(x, y)
plt.show() #函数图像
```
